```python
import tkinter as tk
import os
import webbrowser
import cv2
import time
import pywhatkit
import googlesearch

def open_notepad():
    os.system('notepad.exe')

def take_photo():
    cap = cv2.VideoCapture(0)
    ret, frame = cap.read()
    if ret:
        cv2.imwrite('photo.jpg', frame)
        cap.release()
        cv2.destroyAllWindows()
        status_label.config(text="Photo captured successfully!", fg="green")
    else:
        status_label.config(text="Failed to capture photo.", fg="red")

def play_lofi_music():
    webbrowser.open("https://lofifm.vercel.app/")

def search_on_google():
    query = entry.get()
    url = f"https://www.google.com/search?q={query}"
    webbrowser.open(url)

def start_pomodoro():
    work_time = 25 * 60
    short_break_time = 5 * 60
    long_break_time = 15 * 60
    num_work_sessions = 4
    
    global pomodoro_active
    pomodoro_active = True

    while pomodoro_active and num_work_sessions > 0:
        countdown(work_time, "Work Time")
        if pomodoro_active:
            countdown(short_break_time, "Short Break Time")
            num_work_sessions -= 1
    
    if pomodoro_active:
        countdown(long_break_time, "Long Break Time")

    pomodoro_active = False
    timer_label.config(text="Pomodoro Stopped", fg="red")

def stop_pomodoro():
    global pomodoro_active
    pomodoro_active = False
    timer_label.config(text="Pomodoro Stopped", fg="red")

def countdown(seconds, session_type):
    global pomodoro_active
    while seconds and pomodoro_active:
        mins, secs = divmod(seconds, 60)
        timer_label.config(text=f"{session_type}\n{mins:02d}:{secs:02d}", fg="black")
        root.update()
        time.sleep(1)
        seconds -= 1
    if pomodoro_active:
        timer_label.config(text="Session Complete!", fg="green")
        root.update()
        time.sleep(2)
        timer_label.config(text="")
        root.update()

def calculate():
    expression = entry_calculator.get()
    try:
        result = eval(expression)
        calculator_result.config(text=f"Result: {result}", fg="green")
    except Exception as e:
        calculator_result.config(text="Error", fg="red")


todo_list = []

def add_task():
    task = entry_task.get()
    if task:
        todo_list.append(task)
        update_todo_list()
        entry_task.delete(0, tk.END)

def remove_task(task):
    todo_list.remove(task)
    update_todo_list()

def update_todo_list():
    todo_listbox.delete(0, tk.END)
    for task in todo_list:
        todo_listbox.insert(tk.END, task)
        
def create_indian_flag():
    import numpy as np
    from PIL import Image

    # Define the dimensions of the image
    width = 300
    height = 200

    # Create the Indian flag colors
    saffron = [255, 153, 51]  # RGB value for saffron (#FF9933)
    white = [255, 255, 255]   # RGB value for white (#FFFFFF)
    green = [19, 136, 8]      # RGB value for green (#138808)

    # Create a NumPy array representing the Indian flag
    flag_array = np.zeros((height, width, 3), dtype=np.uint8)

    # Fill the first stripe with saffron color
    flag_array[:height//3, :] = saffron

    # Fill the second stripe with white color
    flag_array[height//3:2*height//3, :] = white

    # Fill the third stripe with green color
    flag_array[2*height//3:, :] = green

    # Convert the NumPy array to a Pillow image
    image = Image.fromarray(flag_array)

    # Display the image using the default image viewer
    image.show()

def send_whatsapp_message():
    pywhatkit.sendwhatmsg_instantly("+919618103439", "Hi", 17, 5)
    




root = tk.Tk()
root.title("Useful Toolbox")
root.geometry("400x700")
root.configure(bg="#f0f0f0") 

# Create and position the buttons
button1 = tk.Button(root, text="Open Notepad", command=open_notepad, padx=10, pady=5, bg="#4caf50", fg="white")
button1.pack(pady=10)

button2 = tk.Button(root, text="Take Photo", command=take_photo, padx=10, pady=5, bg="#2196f3", fg="white")
button2.pack(pady=10)

button3 = tk.Button(root, text="Play Lofi Music", command=play_lofi_music, padx=10, pady=5, bg="#673ab7", fg="white")
button3.pack()

entry = tk.Entry(root, width=30)
entry.pack(pady=10)

button4 = tk.Button(root, text="Search on Google", command=search_on_google, padx=10, pady=5, bg="#ff5722", fg="white")
button4.pack()

button_pomodoro = tk.Button(root, text="Start Pomodoro", command=start_pomodoro, padx=10, pady=5, bg="#ff9800", fg="white")
button_pomodoro.pack(pady=20)

button_stop_pomodoro = tk.Button(root, text="Stop Pomodoro", command=stop_pomodoro, padx=10, pady=5, bg="#e91e63", fg="white")
button_stop_pomodoro.pack(pady=10)

status_label = tk.Label(root, text="", fg="green", font=("Helvetica", 12), bg="#f0f0f0")
status_label.pack()

timer_label = tk.Label(root, text="", font=("Helvetica", 20), bg="#f0f0f0")
timer_label.pack()

# Calculator Section
calculator_label = tk.Label(root, text="Calculator", font=("Helvetica", 16), bg="#f0f0f0")
calculator_label.pack(pady=10)

entry_calculator = tk.Entry(root, width=30)
entry_calculator.pack(pady=5)

calculator_button = tk.Button(root, text="Calculate", command=calculate, padx=10, pady=5, bg="#607d8b", fg="white")
calculator_button.pack()

calculator_result = tk.Label(root, text="", font=("Helvetica", 14), bg="#f0f0f0")
calculator_result.pack()

# To-Do List Section
todo_label = tk.Label(root, text="To-Do List", font=("Helvetica", 16), bg="#f0f0f0")
todo_label.pack(pady=10)

entry_task = tk.Entry(root, width=30)
entry_task.pack(pady=5)

button_add_task = tk.Button(root, text="Add Task", command=add_task, padx=10, pady=5, bg="#009688", fg="white")
button_add_task.pack()

todo_listbox = tk.Listbox(root, selectbackground="#b2dfdb", selectmode=tk.SINGLE, width=40, height=10)
todo_listbox.pack()

button_remove_task = tk.Button(root, text="Remove Task", command=lambda: remove_task(todo_listbox.get(tk.ACTIVE)), padx=10, pady=5, bg="#ff5722", fg="white")
button_remove_task.pack(pady=10)

button_indian_flag = tk.Button(root, text="Create Indian Flag", command=create_indian_flag, padx=10, pady=5, bg="#ffc107", fg="black")
button_indian_flag.pack(pady=10)

button_send_whatsapp = tk.Button(root, text="Send WhatsApp Message", command=send_whatsapp_message, padx=10, pady=5, bg="#2196f3", fg="white")
button_send_whatsapp.pack(pady=10)


# Adjust the placement of existing elements
todo_listbox.place(x=30, y=385)
button_remove_task.place(x=30, y=555)

root.mainloop()
```
